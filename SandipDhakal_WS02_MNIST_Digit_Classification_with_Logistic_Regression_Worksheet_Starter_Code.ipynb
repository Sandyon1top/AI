##  **Some Helper Function:**

### Softmax Function:

```python
import numpy as np

def softmax(z):
    """
    Compute the softmax probabilities for a given input matrix.

    Parameters:
    z (numpy.ndarray): Logits (raw scores) of shape (m, n), where
                       - m is the number of samples.
                       - n is the number of classes.

    Returns:
    numpy.ndarray: Softmax probability matrix of shape (m, n), where
                   each row sums to 1 and represents the probability
                   distribution over classes.

    Notes:
    - The input to softmax is typically computed as: z = XW + b.
    - Uses numerical stabilization by subtracting the max value per row.
    """

    # Numerical stability: subtract max value of each row to prevent overflow
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))

    # Calculate the probabilities
    softmax_probs = exp_z / np.sum(exp_z, axis=1, keepdims=True)

    return softmax_probs
```

### Prediction Function:

```python
def predict_softmax(X, W, b):
    """
    Predict the class labels for a set of samples using the trained softmax model.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d), where n is the number of samples and d is the number of features.
    W (numpy.ndarray): Weight matrix of shape (d, c), where c is the number of classes.
    b (numpy.ndarray): Bias vector of shape (c,).

    Returns:
    numpy.ndarray: Predicted class labels of shape (n,), where each value is the index of the predicted class.
    """

    # Step 1: Compute the linear combination (Z = XW + b)
    z = np.dot(X, W) + b

    # Step 2: Apply softmax to get probability distributions
    probabilities = softmax(z)

    # Step 3: Get the index of the highest probability for each sample
    # axis=1 looks across the columns (the classes)
    predicted_classes = np.argmax(probabilities, axis=1)

    return predicted_classes
```

### Loss Function:

```python
def loss_softmax(y_pred_probs, y_true_onehot):
    """
    Compute the mean cross-entropy loss for a batch of samples.

    Parameters:
    y_pred_probs (numpy.ndarray): Predicted probabilities of shape (n, c),
                                  where n is the number of samples and c is the number of classes.
    y_true_onehot (numpy.ndarray): True labels (one-hot encoded) of shape (n, c).

    Returns:
    float: Mean cross-entropy loss over the batch.
    """
    num_samples = y_true_onehot.shape[0]

    # Clip probabilities to avoid log(0) which leads to infinite loss
    y_pred_probs = np.clip(y_pred_probs, 1e-15, 1 - 1e-15)

    # Compute cross-entropy loss for each sample: -sum(y_true * log(y_pred_probs)) over classes
    # This results in a 1D array of losses, one for each sample in the batch.
    sample_losses = -np.sum(y_true_onehot * np.log(y_pred_probs), axis=1)

    # Return the mean loss for the batch
    return np.mean(sample_losses)

def cost_softmax(X, y_true_onehot, W, b):
    """
    Compute the average softmax regression cost (cross-entropy loss) over all samples.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d), where n is the number of samples and d is the number of features.
    y_true_onehot (numpy.ndarray): True labels (one-hot encoded) of shape (n, c),
                                   where n is the number of samples and c is the number of classes.
    W (numpy.ndarray): Weight matrix of shape (d, c).
    b (numpy.ndarray): Bias vector of shape (c,).

    Returns:
    float: Average softmax cost (cross-entropy loss) over all samples.
    """
    # Step 1: Compute the linear combination (Z = XW + b)
    z = np.dot(X, W) + b

    # Step 2: Apply softmax to get probability distributions
    y_pred_probs = softmax(z)

    # Step 3: Compute the mean cross-entropy loss using the loss_softmax function
    cost = loss_softmax(y_pred_probs, y_true_onehot)

    return cost
```

### Cost Function:

```python
def cost_softmax(X, y, W, b):
    """
    Compute the average softmax regression cost (cross-entropy loss) over all samples.
    """
    m = X.shape[0]

    # 1. Compute the raw scores (Z)
    z = np.dot(X, W) + b

    # 2. Get the probabilities (y_probs) using the softmax function
    # DO NOT use predict_softmax here, as it returns labels (0-9) instead of probabilities.
    y_probs = softmax(z)

    # 3. Calculate the loss for all samples using your loss_softmax function
    # Note: If your loss_softmax is designed for a batch, it returns an array of losses.
    sample_losses = loss_softmax(y_probs, y)

    # 4. Return the average cost
    cost = np.sum(sample_losses) / m
    return cost
```

### Computing Gradients:

```python
def compute_gradient_softmax(X, y, W, b):
    """
    Compute the gradients of the cost function with respect to weights and biases.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d).
    y (numpy.ndarray): True labels (one-hot encoded) of shape (n, c).
    W (numpy.ndarray): Weight matrix of shape (d, c).
    b (numpy.ndarray): Bias vector of shape (c,).

    Returns:
    tuple: Gradients with respect to weights (d, c) and biases (c,).
    """
    m = X.shape[0] # Number of samples

    # 1. Get current predictions (probabilities)
    z = np.dot(X, W) + b
    y_pred = softmax(z)

    # 2. Compute the error (difference between prediction and truth)
    # Since y is one-hot encoded, we can subtract directly
    error = y_pred - y

    # 3. Compute gradients
    # Gradient for weights: (X.T * error) / m
    grad_W = np.dot(X.T, error) / m

    # Gradient for biases: average of the error across samples
    grad_b = np.sum(error, axis=0) / m

    return grad_W, grad_b
```

### Implementing Gradient Descent:

```python
def gradient_descent_softmax(X, y, W, b, alpha, n_iter, show_cost=False):
    """
    Perform gradient descent to optimize the weights and biases.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d).
    y (numpy.ndarray): True labels (one-hot encoded) of shape (n, c).
    W (numpy.ndarray): Weight matrix of shape (d, c).
    b (numpy.ndarray): Bias vector of shape (c,).
    alpha (float): Learning rate.
    n_iter (int): Number of iterations.
    show_cost (bool): Whether to display the cost at intervals.

    Returns:
    tuple: Optimized weights, biases, and cost history.
    """
    cost_history = []

    for i in range(n_iter):
        # Compute gradients
        grad_W, grad_b = compute_gradient_softmax(X, y, W, b)

        # Update weights and biases using the learning rate (alpha)
        W = W - alpha * grad_W
        b = b - alpha * grad_b

        # Calculate current cost and append to history
        current_cost = cost_softmax(X, y, W, b)
        cost_history.append(current_cost)

        # Display cost if requested (every 100 iterations is common)
        if show_cost and i % 100 == 0:
            print(f"Iteration {i}: Cost {current_cost:.4f}")

    return W, b, cost_history
```

## Preparing Dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

def load_and_prepare_mnist(csv_file, test_size=0.2, random_state=42):
    """
    Reads the MNIST CSV file, splits data into train/test sets, and plots one image per class.

    Arguments:
    csv_file (str)       : Path to the CSV file containing MNIST data.
    test_size (float)    : Proportion of the data to use as the test set (default: 0.2).
    random_state (int)   : Random seed for reproducibility (default: 42).

    Returns:
    X_train, X_test, y_train, y_test : Split dataset.
    """

    # Load dataset
    df = pd.read_csv(csv_file)

    # Separate labels and features
    y = df.iloc[:, 0].values  # First column is the label
    X = df.iloc[:, 1:].values  # Remaining columns are pixel values

    # Normalize pixel values (optional but recommended)
    X = X / 255.0  # Scale values between 0 and 1

    # Split data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    # Plot one sample image per class
    plot_sample_images(X, y)

    return X_train, X_test, y_train, y_test

def plot_sample_images(X, y):
    """
    Plots one sample image for each digit class (0-9).

    Arguments:
    X (np.ndarray): Feature matrix containing pixel values.
    y (np.ndarray): Labels corresponding to images.
    """

    plt.figure(figsize=(10, 4))
    unique_classes = np.unique(y)  # Get unique class labels

    for i, digit in enumerate(unique_classes):
        index = np.where(y == digit)[0][0]  # Find first occurrence of the class
        image = X[index].reshape(28, 28)  # Reshape 1D array to 28x28

        plt.subplot(2, 5, i + 1)
        plt.imshow(image, cmap='gray')
        plt.title(f"Digit: {digit}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def loss_softmax(y_pred, y):
    """
    Compute the cross-entropy loss for a batch of samples.
    y_pred: (m, 10) matrix of probabilities
    y: (m, 10) matrix of one-hot encoded labels
    """
    epsilon = 1e-15
    # Ensure both are 2D. If y is (m, 10) and y_pred is (m, 10),
    # the multiplication (*) is element-wise.
    # axis=1 sums the 10 columns into 1 value per sample.
    sample_losses = -np.sum(y * np.log(y_pred + epsilon), axis=1)
    return sample_losses

def cost_softmax(X, y, W, b):
    """
    Compute the average cost over all samples.
    """
    m = X.shape[0]
    # 1. Get raw scores (Z)
    z = np.dot(X, W) + b
    # 2. Get probabilities (Softmax)
    y_pred_probs = softmax(z)

    # 3. Calculate mean of the losses
    cost = np.mean(loss_softmax(y_pred_probs, y))
    return cost

def compute_gradient_softmax(X, y, W, b):
    """
    Compute gradients using the difference between probabilities and one-hot labels.
    """
    m = X.shape[0]
    z = np.dot(X, W) + b
    y_pred = softmax(z)

    # Error is (m, 10) - (m, 10)
    error = y_pred - y

    grad_W = np.dot(X.T, error) / m
    grad_b = np.sum(error, axis=0) / m

    return grad_W, grad_b
```

## **Evaluating the Model:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Evaluation Function
def evaluate_classification(y_true, y_pred):
    """
    Evaluate classification performance using confusion matrix, precision, recall, and F1-score.

    Parameters:
    y_true (numpy.ndarray): True labels
    y_pred (numpy.ndarray): Predicted labels

    Returns:
    tuple: Confusion matrix, precision, recall, F1 score
    """
    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Compute precision, recall, and F1-score
    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')

    return cm, precision, recall, f1
```

```python
def plot_decision_boundary(ax, model, X, y, title):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, cmap='RdBu')
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='RdBu')
    ax.set_title(title)

# Initialize Models
model_lin = LogisticRegression()
model_circ = LogisticRegression()

# Train
model_lin.fit(X_train_lin, y_train_lin)
model_circ.fit(X_train_circ, y_train_circ)

# Create the 2x2 Plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
plot_decision_boundary(axes[0,0], model_lin, X_train_lin, y_train_lin, "Linear (Train)")
plot_decision_boundary(axes[0,1], model_lin, X_test_lin, y_test_lin, "Linear (Test)")
plot_decision_boundary(axes[1,0], model_circ, X_train_circ, y_train_circ, "Non-Linear (Train)")
plot_decision_boundary(axes[1,1], model_circ, X_test_circ, y_test_circ, "Non-Linear (Test)")
plt.tight_layout()
plt.show()
```

# Linear Seperability and Logistic Regression:

Question 2: Interpretation of the Decision Boundary Plots
Observation: The Logistic Regression model performed significantly better on the Linearly Separable (Blobs) dataset compared to the Non-Linearly Separable (Circles) dataset.

Interpretation:

Linearly Separable Data: Logistic Regression is a linear classifier, meaning its decision boundary is a straight line (a hyperplane in higher dimensions). Since the "Blobs" dataset has two distinct groups that can be separated by a single straight line, the model achieved near-perfect accuracy on both training and testing sets.

Non-Linearly Separable Data: For the "Circles" dataset, the classes are arranged as an inner circle and an outer ring. Because the model is mathematically constrained to drawing a straight line, it cannot "curve" around the inner group. This results in the model essentially splitting the circle in half, misclassifying about 50% of the data. To solve this, one would need non-linear features (like polynomial terms) or a different model like a Support Vector Machine (SVM) with a RBF kernel.

A step size too large made the boundary look jagged, while a step size too small slowed down the rendering process significantly.Numerical Stability: Implementing the Softmax function from scratch required careful handling of large exponents. Without subtracting the maximum value from the logits ($z$), the np.exp() function would frequently result in NaN (Not a Number) errors due to numerical overflow.Convergence Monitoring: Finding the right balance for the learning rate ($\alpha$) and the number of iterations was a trial-and-error process to ensure the cost function decreased steadily without oscillating.
