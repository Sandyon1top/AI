##Task 1 – Unit Converter

```python
def convert_length(value, direction):
    """
    Converts between meters and feet.
    """
    if direction == 'm_to_ft':
        return value * 3.28084
    elif direction == 'ft_to_m':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction for length conversion.")

def convert_weight(value, direction):
    """
    Converts between kilograms and pounds.
    """
    if direction == 'kg_to_lbs':
        return value * 2.20462
    elif direction == 'lbs_to_kg':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction for weight conversion.")

def convert_volume(value, direction):
    """
    Converts between liters and gallons.
    """
    if direction == 'l_to_gal':
        return value * 0.264172
    elif direction == 'gal_to_l':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction for volume conversion.")

def unit_converter():
    """Drives the unit converter program."""
    print("Choose conversion type:")
    print("1. Length  2. Weight  3. Volume")
    try:
        choice = int(input("Enter 1, 2, or 3: "))
        value = float(input("Enter value: "))
        if choice == 1:
            direction = input("Enter m_to_ft or ft_to_m: ")
            result = convert_length(value, direction)
        elif choice == 2:
            direction = input("Enter kg_to_lbs or lbs_to_kg: ")
            result = convert_weight(value, direction)
        elif choice == 3:
            direction = input("Enter l_to_gal or gal_to_l: ")
            result = convert_volume(value, direction)
        else:
            raise ValueError("Invalid choice.")
        print(f"Result: {result:.4f}")
    except ValueError as e:
        print(f"Error: {e}")

unit_converter()
```

Task 2 – Math Operations on a List

```python
def compute_sum(numbers):
    """Returns the sum of a list of numbers."""
    return sum(numbers)

def compute_average(numbers):
    """Returns the average of a list of numbers."""
    if not numbers:
        raise ValueError("List is empty.")
    return sum(numbers) / len(numbers)

def compute_maximum(numbers):
    """Returns the maximum value in a list."""
    if not numbers:
        raise ValueError("List is empty.")
    return max(numbers)

def compute_minimum(numbers):
    """Returns the minimum value in a list."""
    if not numbers:
        raise ValueError("List is empty.")
    return min(numbers)

def math_operations():
    """Drives the math operations program."""
    print("Choose an operation: 1. Sum  2. Average  3. Maximum  4. Minimum")
    try:
        choice = int(input("Enter 1-4: "))
        raw = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw.split()]
        if not numbers:
            raise ValueError("No numbers entered.")
        ops = {1: compute_sum, 2: compute_average, 3: compute_maximum, 4: compute_minimum}
        if choice not in ops:
            raise ValueError("Invalid choice.")
        print(f"Result: {ops[choice](numbers)}")
    except ValueError as e:
        print(f"Error: {e}")

math_operations()
```

## 4.3 Nested Lists

```python
def flatten(lst):
    """Flattens a one-level nested list."""
    return [item for sublist in lst for item in sublist]

# Example usage:
print(f"[[1,2],[3,4],[5]] -> {flatten([[1,2],[3,4],[5]])}")
```

```python
def access_nested_element(lst, indices):
    """Returns the element at the position given by indices."""
    result = lst
    for i in indices:
        result = result[i]
    return result

# Example usage:
lst_nested = [[1,2,3],[4,5,6],[7,8,9]]
indices_nested = [1,2]
print(f"lst={lst_nested}, indices={indices_nested} -> {access_nested_element(lst_nested, indices_nested)}")
```

```python
def sum_nested(lst):
    """Recursively sums all elements in a nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

# Example usage:
print(f"[[1,2],[3,[4,5]],6] -> {sum_nested([[1,2],[3,[4,5]],6])}")
```

```python
def remove_element(lst, elem):
    """Removes all occurrences of elem from a nested list."""
    return [
        remove_element(item, elem) if isinstance(item, list) else item
        for item in lst if item != elem
    ]

# Example usage:
print(f"[[1,2],[3,2],[4,5]], elem=2 -> {remove_element([[1,2],[3,2],[4,5]], 2)}")
```

```python
def find_max(lst):
    """Finds the maximum element in a nested list of any depth."""
    flat = []
    def helper(l):
        for item in l:
            helper(item) if isinstance(item, list) else flat.append(item)
    helper(lst)
    return max(flat)

# Example usage:
print(f"[[1,2],[3,[4,5]],6] -> {find_max([[1,2],[3,[4,5]],6])}")
```

```python
def count_occurrences(lst, elem):
    """Counts occurrences of elem in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

# Example usage:
print(f"[[1,2],[2,3],[2,4]], elem=2 -> {count_occurrences([[1,2],[2,3],[2,4]], 2)}")
```

```python
def deep_flatten(lst):
    """Flattens a deeply nested list to any depth."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

# Example usage:
print(f"[[[1,2],[3,4]],[[5,6],[7,8]]] -> {deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]])}")
```

```python
def average_nested(lst):
    """Returns the average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

# Example usage:
print(f"[[1,2],[3,4],[5,6]] -> {average_nested([[1,2],[3,4],[5,6]])}")
```

### Problem 4 – Matrix Operations

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

print(f"Matrix A:\n{A}\n")
print(f"Matrix B:\n{B}\n")

# 1. Prove A * A_inv = I
A_inv = np.linalg.inv(A)
print(f"Inverse of A:\n{A_inv}\n")
print(f"A * A_inv (should be identity matrix):\n{np.round(np.dot(A, A_inv))}\n")

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print(f"AB:\n{AB}\n")
print(f"BA:\n{BA}\n")
print(f"AB == BA?: {np.array_equal(AB, BA)}\n")

# 3. Prove (AB)^T = B^T * A^T
print(f"(AB)^T:\n{np.dot(A, B).T}\n")
print(f"B^T * A^T:\n{np.dot(B.T, A.T)}\n")
print(f"(AB)^T == B^T * A^T?: {np.array_equal(np.dot(A, B).T, np.dot(B.T, A.T))}\n")

# Solve system of linear equations
# 2x - 3y + z = -1
# x  - y + 2z = -3
# 3x + y  - z =  9
A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]], dtype=float)
b_sys = np.array([-1, -3, 9], dtype=float)

print(f"System Matrix A:\n{A_sys}\n")
print(f"System Vector b:\n{b_sys}\n")

# Method 1: Inverse method
X_inv = np.dot(np.linalg.inv(A_sys), b_sys)
print(f"Solution using inverse method: x={X_inv[0]:.2f}, y={X_inv[1]:.2f}, z={X_inv[2]:.2f}\n")

# Method 2: np.linalg.solve
X_solve = np.linalg.solve(A_sys, b_sys)
print(f"Solution using np.linalg.solve: x={X_solve[0]:.2f}, y={X_solve[1]:.2f}, z={X_solve[2]:.2f}\n")
```

## 10.2 Experiment: How Fast is NumPy?

```python
import numpy as np
import time

SIZE = 1_000_000

list_a = list(range(SIZE))
list_b = list(range(SIZE))
np_a = np.arange(SIZE, dtype=float)
np_b = np.arange(SIZE, dtype=float)

print(f"Comparing operations for {SIZE} elements:\n")

# 1. Element-wise Addition
t = time.time()
_ = [list_a[i] + list_b[i] for i in range(SIZE)]
print(f"List add:   {time.time()-t:.4f}s")
t = time.time()
_ = np_a + np_b
print(f"NumPy add:  {time.time()-t:.4f}s\n")

# 2. Element-wise Multiplication
t = time.time()
_ = [list_a[i] * list_b[i] for i in range(SIZE)]
print(f"List mul:   {time.time()-t:.4f}s")
t = time.time()
_ = np_a * np_b
print(f"NumPy mul:  {time.time()-t:.4f}s\n")

# 3. Dot Product
t = time.time()
_ = sum(list_a[i]*list_b[i] for i in range(SIZE))
print(f"List dot:   {time.time()-t:.4f}s")
t = time.time()
_ = np.dot(np_a, np_b)
print(f"NumPy dot:  {time.time()-t:.4f}s\n")

# 4. Matrix Multiplication
MAT = 100
# For performance comparison, use a smaller matrix size if MAT=1000 causes OOM or very long execution.
# If MAT=1000 is intended for the prompt, comment out or adjust as necessary for Colab environment.
# The example in the original text used MAT=1000 but the time for list matmul was '%.2f's which implies a longer run.
mat_np = np.random.rand(MAT, MAT)
mat_list = mat_np.tolist()

print(f"Comparing matrix multiplication for {MAT}x{MAT} matrices:\n")
t = time.time()
_ = [[sum(mat_list[r][k]*mat_list[k][c] for k in range(MAT)) for c in range(MAT)] for r in range(MAT)]
print(f"List matmul:   {time.time()-t:.4f}s")

t = time.time()
_ = np.dot(mat_np, mat_np)
print(f"NumPy matmul:  {time.time()-t:.4f}s\n")
```
